#### Magnitud y dirección de una relación lineal entre 2 variables

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importamos la data desde un archivo CSV

personas = pd.read_csv("../datos/personas.csv")
personas
```

```python
# Realizamos un gráfico para observar mejor nuestra data

personas.plot()
```

##### Relación: Altura y Peso

```python
# Creamos un gráfica de dispersión de nuestra data
# Usamos la librería matplotlib

plt.scatter(personas["altura"], personas["peso"])
plt.xlabel("Altura [cm]")
plt.ylabel("Peso [kg]")
plt.grid(True)
plt.show()
```

##### Relación: Ingreso y Horas Trabajadas

```python
# Creamos un gráfica de dispersión de nuestra data
# Usamos la librería matplotlib

plt.scatter(personas["ingreso"], personas["horas_trabajadas"])
plt.xlabel("Ingreso [$]")
plt.ylabel("Horas Trabajadas")
plt.grid(True)
plt.show()
```

##### Relación: Ingreso y Ausencias

```python
# Creamos un gráfica de dispersión de nuestra data
# Usamos la librería matplotlib

plt.scatter(personas["ingreso"], personas["ausencias"])
plt.xlabel("Ingreso [$]")
plt.ylabel("Ausencias")
plt.grid(True)
plt.show()
```

##### Relación: Ingreso y Peso de una Persona

```python
# Creamos un gráfica de dispersión de nuestra data
# Usamos la librería matplotlib

plt.scatter(personas["ingreso"], personas["peso"])
plt.xlabel("Ingreso [$]")
plt.ylabel("Peso [kg]")
plt.grid(True)
plt.show()
```

#### Matriz de Correlación

```python
# Calculamos la matriz de correlación de nuestro DataFrame

matriz = personas.corr()
matriz

# Los valores van desde: -1 (existe una realción fuerte) - 0 - 1 (existe una relación fuerte)
```

```python
# Mostrar nuestro gráfico de calor de las correlaciones

# El parámetro "bwr" srive para modificar los colores del mapa
# El primer valor "b" es para el -1
# El segundo valor "w" es para el 0
# El tercer valor "r" es para el 1

# El parámetro "vmin" sirve para mostrar los valores mínimos de la figura
# El parámetro "vmax" srive para mostrar los valores máximos de la figura

plt.matshow(matriz, cmap = "bwr", vmin = -1, vmax = 1)
plt.colorbar() # Poner un barra de color 

# .xticks(# de puntos o ticks, valores o string a mostrar, rotation = se indican los grados para rotar el string/texto)

plt.xticks(range(5), personas.columns, rotation = 90) # Para cambiar el texto que muestran los puntos de la figura en el eje x
plt.yticks(range(5), personas.columns) # Para cambiar el texto que muestran los puntos de la figura en el eje y

# Vamos a llenar nuestro gráfico con texto correspondiente a cada punto

for i in range(len(matriz.columns)):
    for j in range(len(matriz.columns)):
        plt.text(i, j, round(matriz.iloc[i, j], 2), ha = "center", va = "center")

plt.show()
```
