## CREACIÓN Y MANIPULACIÓN DE DATFRAMES

```python
import pandas as pd
import numpy as np
```

#### Creación de un DataFrame

```python
# Definir los datos para el DataFrame

nombre_paises = ["China", "India", "Estados Unidos", "Indonesia", "Pakistán",
                 "Brasil", "Nigeria", "Bangladesh", "Rusia", "México"]

encabezados = ["Poblacion", "Porcentaje"]

datos = [[1439, 18.47],
        [1380, 17.70],
        [331, 4.25],
        [273, 3.51], 
        [220, 2.83],
        [212, 2.73], 
        [206, 2.64],
        [164, 2.11],
        [145, 1.87],
        [128, 1.65]]
```

```python
# Creación de un nuevo DataFrame normal

paises = pd.DataFrame(datos)
paises
```

```python
# Creación de un nuevo DataFrame con índices

paises = pd.DataFrame(data = datos, index = nombre_paises)
paises
```

```python
# Creación de un nuevo DataFrame con índices y columnas (encabezados)

paises = pd.DataFrame(data = datos, index = nombre_paises, columns = encabezados)
paises
```

```python
# Crear un DataFrame a partir de un diccionario sin indices

datos = {"China": [1439, 18.47],
         "India": [1380, 17.70],
         "Estados Unidos": [331, 4.25],
         "Indonesia": [273, 3.51], 
         "Pakistán": [220, 2.83],
         "Brasil": [212, 2.73], 
         "Nigeria": [206, 2.64],
         "Bangladesh": [164, 2.11],
         "Rusia": [145, 1.87],
         "México": [128, 1.65]}

paises = pd.DataFrame(data = datos)
paises
```

```python
# Crear un DataFrame a partir de un diccionario con indices
# Cada clave del diccionario equivale una columna con sus respectivos valores

datos = {"China": [1439, 18.47],
         "India": [1380, 17.70],
         "Estados Unidos": [331, 4.25],
         "Indonesia": [273, 3.51], 
         "Pakistán": [220, 2.83],
         "Brasil": [212, 2.73], 
         "Nigeria": [206, 2.64],
         "Bangladesh": [164, 2.11],
         "Rusia": [145, 1.87],
         "México": [128, 1.65]}

paises = pd.DataFrame(data = datos, index = encabezados)
paises
```

```python
# Obtener un transpuesto del DataFrame original

paises = paises.T
paises

#Otra forma es utilizar el método 
# "paises.transpose()"
```

#### Acceder a los elementos de un DataFrame

```python
# Acceder a una columna del DataFrame

paises.Poblacion
```

```python
# Otra forma de acceder a las columnas del DataFrame

paises["Poblacion"]
```

```python
# Acceder a varias columnas del DataFrame

paises[["Poblacion", "Porcentaje"]]
```

```python
# Accder a un celda en particular del DataFrame

valor_1 = paises["Poblacion"][0] #Accediento a un valor
valor_2 = paises["Poblacion"]["China"]
valor_3 = paises.Poblacion["China"]
valor_4 = paises["Poblacion"][:] #Accediendo a toda la fila/celda

print(valor_1)
print(valor_2)
print(valor_3, end = "\n\n")
print(valor_4)
```

```python
# Acceder a un renglon/fila mediante su posición, la posicón es respecto al índice/nombre

paises.iloc[2]
```

```python
# Acceder a un renglon mediante su posición, la posicón es respecto al índice/nombre

filtro = np.where(paises.index == "Brasil")
paises.iloc[paises.index == "Brasil"][["Porcentaje"]]
```

```python
# Acceder a un renglon mediante su nombre/índice

paises.loc["China"]
```

```python
# Acceder a un renglon/fila mediante su nombre/índice, para este caso un rango

paises.loc["China": "Brasil", "Porcentaje"]
```

```python
# Obtener los primeros valores del DataFrame

paises.head(n = 3)
```

```python
# Ordenar los valores del DataFrame mediante mis índices
# Hay que especificar la columna con "by", y con ascending puedo decirle que lo haga de manera de menor a mayor

paises.sort_values(by = ["Porcentaje"], ascending = True)
```

```python
# Ordenar los valores del DataFrame mediante los índices
# Con ascending se ordena de menor a mayor

paises.sort_index(ascending = True)
```

#### Agregar y borrar renglones y columnas

```python
# Creamos el nuevo renglon

tasa_fertilidad = [1.7, 2.2, 1.8, 2.3, 3.6, 1.7, 5.4, 2.1, 1.8, 2.1]
```

```python
# Agregamos el nuevo renglon

paises["Tasa_Fertilidad"] = tasa_fertilidad
paises
```

```python
# Borrar una columna del DataFrame
# Devuelve la columna que se va a borrar

paises.pop("Tasa_Fertilidad")
```

```python
paises
```

```python
# Agregamos de nuevo la columna del DataFrame

paises["Tasa_Fertilidad"] = tasa_fertilidad
paises
```

```python
# Otra forma de borrar una columna del DataFrame
# Este método no devuelve la columna que se va a borrar

del paises["Tasa_Fertilidad"]
paises
```

```python
# Agregamos de nuevo la columna del DataFrame

paises["Tasa_Fertilidad"] = tasa_fertilidad
paises
```

```python
# Otra forma de borrar columna o reglones del DataFrame
# Regresa la vista del DataFrame sin modificar el origibal
# "axis = 1" borra la columna, "axis = 0" borra la fila

paises.drop(["Rusia", "México"], axis = 0)
```

```python
paises
```

```python
# Creamos el nuevo renglon para agregar

renglon = pd.Series(name = "Japon", data = [126, 1.62], index = ["Poblacion", "Porcentaje"])
renglon
```

```python
# Agregar un nuevo renglon al DataFrame

paises = paises.append(renglon)
paises
```

```python
# Modificar el valor de una celda en concreto

paises["Tasa_Fertilidad"]["Japon"] = 2.8
paises
```
