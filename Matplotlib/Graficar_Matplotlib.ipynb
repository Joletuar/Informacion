## Gráficas con Matplotlib

#### Creación de Gráficas de Líneas (plot)

```python
import matplotlib.pyplot as plt
```

```python
# Creamos nuestras etiquetas para el eje x

eje_x = ["a", "b", "c", "d", "e"]
```

```python
# Creamos los datos a graficar

datos_1 = [10, 20, 30, 40, 50]
datos_2 = [60, 70, 80, 90, 100]
```

```python
# Realizamos nuestra gráfica con (plot)

# Con el "plot" estamos agregando una serie de datos pero no la estamos mostrando

plt.plot(eje_x, datos_1)
plt.plot(eje_x, datos_2)

# Con el "show()" le decimos que muestre la serie de datos que se encuentran agregadas
# Tambien se encarga de limpiar nuestro lienzo

plt.show() 
```

```python
# Realizamos nuestra gráfica con (plot)
# Graficamos y limpiamos el lienzo para mostrar la siguiente

plt.plot(eje_x, datos_1)
plt.show() 

plt.plot(eje_x, datos_2)
plt.show() 
```

#### Generar Datos a Graficar

```python
import numpy as np
```

```python
# Especificamos nuestros datos a graficar

anios = ["2011", "2012", "2014", "2014", "2015", 
         "2016", "2017", "2018", "2019", "2020"]

puntajes_nivel_1 = np.random.randint( low = 1, high = 100, size=10)
puntajes_nivel_2 = np.random.randint( low = 25, high = 100, size=10)
puntajes_nivel_3 = np.random.randint( low = 50, high = 100, size=10)
puntajes_nivel_4 = np.random.randint( low = 75, high = 100, size=10)
puntajes_nivel_5 = np.random.randint( low = 90, high = 100, size=10)
```

#### Personalización de Gráficas

```python
# Introducir series de datos

plt.plot(anios, puntajes_nivel_1)
plt.plot(anios, puntajes_nivel_2)
plt.plot(anios, puntajes_nivel_3)
plt.plot(anios, puntajes_nivel_4)
plt.plot(anios, puntajes_nivel_5)

# Mostramos la gráfica

plt.show()
```

```python
# Modificar los colores de las gráficas

# Introducimos nuestras series/data y especificamos el color de cada línea

plt.plot(anios, puntajes_nivel_1, color = "black")
plt.plot(anios, puntajes_nivel_2, color = "blue")
plt.plot(anios, puntajes_nivel_3, color = "yellow")
plt.plot(anios, puntajes_nivel_4, color = "red")
plt.plot(anios, puntajes_nivel_5, color = "gray")

# Mostramos la gráfica

plt.show()
```

```python
# Modificar los marcadores/puntos de la gráficas

# Introducimos nuestras series/data y especificamos el tipo de marcador de cada línea

plt.plot(anios, puntajes_nivel_1, marker = "*")
plt.plot(anios, puntajes_nivel_2, marker = "<")
plt.plot(anios, puntajes_nivel_3, marker = ">")
plt.plot(anios, puntajes_nivel_4, marker = "+")
plt.plot(anios, puntajes_nivel_5, marker = ".")

# Mostramos la gráfica

plt.show()
```

```python
# Modificar los tipos de líneas de la gráficas

# Introducimos nuestras series/data y especificamos el tipo de linea de cada gráfica

plt.plot(anios, puntajes_nivel_1, linestyle = "-")
plt.plot(anios, puntajes_nivel_2, linestyle = "-.")
plt.plot(anios, puntajes_nivel_4, linestyle = ":")
plt.plot(anios, puntajes_nivel_3, linestyle = "--")
plt.plot(anios, puntajes_nivel_5, linestyle = "-.")

# Mostramos la gráfica

plt.show()
```

```python
# Activar leyendas e introducir etiquetas

# Especificamos las etiquetas para cada serie

plt.plot(anios, puntajes_nivel_1, label = "Nivel_1")
plt.plot(anios, puntajes_nivel_2, label = "Nivel_2")
plt.plot(anios, puntajes_nivel_3, label = "Nivel_3")
plt.plot(anios, puntajes_nivel_4, label = "Nivel_4")
plt.plot(anios, puntajes_nivel_5, label = "Nivel_5")

# Decimos que se activen las etiquetas/leyendas que especificamos

plt.legend()

# Mostramos la gráfica

plt.show()
```

```python
# Título de la gráfica

# Introducimos nuestras series/data

plt.plot(anios, puntajes_nivel_1)
plt.plot(anios, puntajes_nivel_2)
plt.plot(anios, puntajes_nivel_3)
plt.plot(anios, puntajes_nivel_4)
plt.plot(anios, puntajes_nivel_5)

# Activamos un título para el gráfico

plt.title("Notas de Estudiantes")

# Mostramos el gráfico

plt.show()
```

```python
# Etiqueta de los ejes

# Introducimos nuestras series/data

plt.plot(anios, puntajes_nivel_1)
plt.plot(anios, puntajes_nivel_2)
plt.plot(anios, puntajes_nivel_3)
plt.plot(anios, puntajes_nivel_4)
plt.plot(anios, puntajes_nivel_5)

# Poniendo etiquetas a los ejes

# Para el eje x
plt.xlabel("Años")
# Para el eje y
plt.ylabel("Puntaje")

# Mostramos el gráfico

plt.show()
```

```python
# Modificar los rangos/ticks de la gráfica

# Introducimos nuestras series/data

plt.plot(anios, puntajes_nivel_1)
plt.plot(anios, puntajes_nivel_2)
plt.plot(anios, puntajes_nivel_3)
plt.plot(anios, puntajes_nivel_4)
plt.plot(anios, puntajes_nivel_5)

# Especificamos el rango para los ticks del eje y

plt.yticks(np.arange(0, 101, 10))

# Mostramos el gráfico

plt.show()
```

```python
# Mostrar las regillas/grids mayores

# Introducimos nuestras series/data

plt.plot(anios, puntajes_nivel_1)
plt.plot(anios, puntajes_nivel_2)
plt.plot(anios, puntajes_nivel_3)
plt.plot(anios, puntajes_nivel_4)
plt.plot(anios, puntajes_nivel_5)

# Habilitamos la regilla/grid

plt.grid()

# Mostramos el gráfico

plt.show()
```

```python
# Mostrar las regillas/grids menores

# Introducimos nuestras series/data

plt.plot(anios, puntajes_nivel_1)
plt.plot(anios, puntajes_nivel_2)
plt.plot(anios, puntajes_nivel_3)
plt.plot(anios, puntajes_nivel_4)
plt.plot(anios, puntajes_nivel_5)

# Habilitamos las regillas/grids

plt.minorticks_on()

# Mostramos el gráfico

plt.show()
```

#### Figuras y Paneles

###### Cada figura tiene distintos paneles donde en cada panel puede ir uno varias gráficas

```python
# Creamos nuestro objeto 

fig, panel = plt.subplots()

# Creamos las funciones a graficar

x = np.linspace(-2*np.pi, 2*np.pi, 50)
y = np.sin(x)

# Realizamos la gráfica en el panel

panel.plot(x, y, color = "red", linestyle = "--", 
           marker = "*", linewidth = 0.5, label = "Seno")
panel.grid()
panel.minorticks_on()
panel.legend()
```

#### Figura con varios Paneles

```python
# Crear una figura con 2 paneles horizontales

# Con el método subplots(# de filas, # de columnas)

fig, (panel_1, panel_2) = plt.subplots(2, 1)

# Panel #1

# Creamos la primera función a graficar

x1 = np.linspace(-2*np.pi, 2*np.pi, 50)
y1 = np.cos(x1)

# Realizamos la gráfica en el panel

panel_1.plot(x1, y1, color = "black", linestyle = "--", 
           marker = "*", linewidth = 0.5, label = "Coseno")
panel_1.grid()
panel_1.minorticks_on()
panel_1.legend()
panel_1.set_title("Función Coseno")
panel_1.set_xlabel("Eje x")

# Panel #2

# Creamos segunda función a graficar

x2 = np.linspace(-2*np.pi, 2*np.pi, 50)
y2 = np.sin(x2)

# Realizamos la gráfica en el panel

panel_2.plot(x2, y2, color = "red", linestyle = "--", 
           marker = "*", linewidth = 0.5, label = "Seno")
panel_2.minorticks_on()
panel_2.legend()
panel_2.set_title("Función Seno")
panel_2.set_ylabel("Eje y")

# Ajustar el espacio entre los paneles para que se visualice correctamente el título

fig.tight_layout()
```

```python
# Crear una figura con 2 paneles verticales

# Con el método subplots(# de filas, # de columnas)
# Con el parámetro "constrained_layout = True" evitamos que se amontonen los ejes

fig, (panel_1, panel_2) = plt.subplots(1, 2, constrained_layout = True)

# Panel #1

# Creamos la primera función a graficar

x1 = np.linspace(-2*np.pi, 2*np.pi, 50)
y1 = np.cos(x1)

# Realizamos la gráfica en el panel

panel_1.plot(x1, y1, color = "black", linestyle = "--", 
           marker = "*", linewidth = 0.5, label = "Coseno")
panel_1.grid()
panel_1.minorticks_on()
panel_1.legend()
panel_1.set_title("Función Coseno")
panel_1.set_xlabel("Eje x")

# Panel #2

# Creamos segunda función a graficar

x2 = np.linspace(-2*np.pi, 2*np.pi, 50)
y2 = np.sin(x2)

# Realizamos la gráfica en el panel

panel_2.plot(x2, y2, color = "red", linestyle = "--", 
           marker = "*", linewidth = 0.5, label = "Seno")
panel_2.minorticks_on()
panel_2.legend()
panel_2.set_title("Función Seno")
panel_2.set(xlabel = "Eje x", ylabel = "Eje y") # Forma rápida de dar nombres a los ejes con el método ".set(xlabel = "nombre", ylabel = "nombre")"

plt.show()
```

```python
# Crear una figura con varios paneles

# En este caso paneles será una lista de con los diferentes subplots

fig, paneles = plt.subplots(2,2, constrained_layout = True)

# Creamos las funciones

x1 = np.linspace(-2*np.pi, 2*np.pi, 50)
y1 = np.cos(x1)
paneles[0,0].plot(x1, y1)
paneles[0,0].set_title("Subplot 1") # Damos un título solamente al primer subplot

x2 = np.linspace(-2*np.pi, 2*np.pi, 50)
y2 = np.sin(x2)
paneles[0,1].plot(x2, y2, color = "red")
paneles[0,0].set_title("Subplot 2")

x3 = np.arange(1, 50, 2)
y3 = x3*4 -5
paneles[1,0].plot(x3, y3, color = "black")
paneles[0,0].set_title("Subplot 3")

x4 = np.linspace(1, 100, 50)
y4 = np.sqrt(x4)
paneles[1,1].plot(x4, y4, color = "green")
paneles[0,0].set_title("Subplot 4")

# Iteramos entre todos los ejes y le asignamos un nombre/label a cada eje de los 4 subplots

for label_axe in fig.get_axes():
    # label_axe.set(xlabel = "Eje x", ylabel = "Eje y")
    label_axe.set_xlabel("Eje x")
    label_axe.set_ylabel("Eje y")

# Damos un título a la figura en general

fig.suptitle("Figura con 4 paneles") 

# Compartir los ejes de cada columna y de cada fila

for label_axe in fig.get_axes():
    label_axe.label_outer()
    
# Aumentar el tamaño de la figura

fig.set_size_inches(10,10)
```

#### Figuras y subfiguras

```python
# Creamos un objeto de clase figura, creamos la figura principal

fig = plt.figure(figsize = (10,6), constrained_layout = True)

# Creamos las subfiguras que irán dentro de la figura principal
# Con en el parámetro "wspace = #valor" definimos la separación horizontal entre figuras
# Con en el parámetro "hspace = #valor" definimos la separación vertical entre figuras

subfig_1, subfig_2 = fig.subfigures(1,2, wspace = 0.015)

# Creamos 2 paneles para la primera subfigura

paneles_1 = subfig_1.subplots(2, 1)

# Creamos las funciones a graficar

x1 = np.linspace(-2*np.pi, 2*np.pi, 50)
y1 = np.cos(x1)

x2 = np.linspace(-2*np.pi, 2*np.pi, 50)
y2 = np.sin(x2)

# Graficamos las funciones en la figura 1 con diferentes características

paneles_1[0].plot(x1, y1, color = "red", linestyle = "--", label = "Coseno")
paneles_1[0].legend()
paneles_1[0].set_title("Función Coseno")

paneles_1[1].plot(x2, y2, color = "black", label = "Seno")
paneles_1[1].legend()
paneles_1[1].set_title("Función Seno")

subfig_1.suptitle("$ Figura 1/Hijo $")

# Creamos 4 paneles para la segunda subfigura

paneles_2 = subfig_2.subplots(2, 2)

# Creamos las funciones a graficar

x1 = np.linspace(-2*np.pi, 2*np.pi, 50)
y1 = np.cos(x1)

x2 = np.linspace(-2*np.pi, 2*np.pi, 50)
y2 = np.sin(x2)

x3 = np.arange(1, 50, 2)
y3 = x3*4 -5

x4 = np.linspace(1, 100, 50)
y4 = np.sqrt(x4)

# Graficamos las funciones en la figura 1 con diferentes características

paneles_2[0,0].plot(x1, y1, color = "red", linestyle = "--", label = "Coseno")
paneles_2[0,0].legend()
paneles_2[0,0].set_title("Función Coseno")

paneles_2[0,1].plot(x2, y2, color = "black", label = "Seno")
paneles_2[0,1].legend()
paneles_2[0,1].set_title("Función Seno")

paneles_2[1,0].plot(x3, y3, color = "blue", label = "Lineal")
paneles_2[1,0].legend()
paneles_2[1,0].set_title("Función Lineal")

paneles_2[1,1].plot(x4, y4, color = "green", label = "Raíz")
paneles_2[1,1].legend()
paneles_2[1,1].set_title("Función Raíz")

subfig_2.suptitle("$ Figura 2/Hijo $")

# Modificamos el título general de la figura padre

fig.suptitle("# Figura General/Padre #")
```
